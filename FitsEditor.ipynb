```python
# imports
import ccdproc

from astropy.nddata import CCDData
from astropy.utils.exceptions import AstropyWarning
```

```python
# definitions
initial_dir = "TestData"
final_dir = "FixedData"
```

```python
# collect all files
InitialFits = ccdproc.ImageFileCollection(initial_dir)
```

```python
IniFits = []
for file in InitialFits.files_filtered(include_path=True):
    fits = CCDData.read(file, unit="adu")
    IniFits.append({"FILE": file, "FITS": fits})

print(IniFits)
```

```python
# add bunit
for file in IniFits:
    file["FITS"].meta["BUNIT"] = "adu"
    print(f"added bunit to {file['FILE']}")
```

```python
for file in IniFits:
    imtype = "Light"
    if "Flat" in file["FILE"]: imtype = "Flat"
    file["FITS"].meta["IMTYPE"] = imtype
    print(f"{imtype} {file['FILE']}")
```

```python
for file in IniFits:
    file["FITS"].write(file["FILE"], overwrite=True)
```
